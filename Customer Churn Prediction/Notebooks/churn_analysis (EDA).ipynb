## Data Cleaning

1. Create a copy of base data for manipulation and processingFor the fetures withless missing values can use regression to predivt the missing values or fill with the mean of the values present, depending on the feture

```python
telco_data = telco_base_data.copy()
```

2. Total charges should be numeric amount. let's convert it into numerical data type

```python
telco_data['TotalCharges'] = pd.to_numeric(telco_data['TotalCharges'], errors = 'coerce')
telco_data.isnull().sum()
```

3. As we can see there are 11 missing values in TotalCharges column. Let's check these record

```python
telco_data.loc[telco_data['TotalCharges'].isnull()== True]
```

4. Missing Value Treatment

Since the % of these records compared to total dataset  is very low 0.15%, it is safe to ignore them dor further processing

```python
# Removing the missing values
telco_data.dropna(how = 'any', inplace= True)
```

5. Divide customers into bins cbased on tenure

```python
# Get the max Tenure 
print(telco_data['tenure'].max())
```

```python
#  Group the tenure in bins of 12 month
labels = ['{0}-{1}'.format(i, i+11) for i in range(1, 72,12)]
telco_data['tenure_group'] = pd.cut(telco_data['tenure'], range(1,80,12), right =False, labels = labels)
```

```python
telco_data['tenure_group'].value_counts()
```

6. Remove the columns not required for processing 

```python
#  drop coumn customer id and tenure
telco_data.drop(columns=['customerID','tenure'], axis=1, inplace = True)
telco_data.head()
```

## Data Exploration

1. Plot distribution of individual predictor by churn

### Univariate Analysis

```python
for i, predictor in enumerate(telco_data.drop(columns=['Churn','TotalCharges','MonthlyCharges'])):
    plt.figure(i)
    sns.countplot(data = telco_data, x= predictor, hue= 'Churn')
    plt.show()
```

2. Convert the target variable 'Churn' in binary numerical variable

```python
telco_data['Churn'] = np.where(telco_data['Churn']== 'Yes',1,0)
```

```python
telco_data.head()
```

3. Convert all the categorical variable into dummy variables

```python
telco_data_dummies = pd.get_dummies(telco_data)
```

```python
telco_data_dummies.head()
```

Relationship between Monthly charges and Total charges

```python
sns.lmplot(data = telco_data_dummies, x = 'MonthlyCharges', y = 'TotalCharges', fit_reg=False)
plt.show()
```

Total charges increas as Monthly Charges increase - as expected

Churn by Monthly Charges and total charges

```python
mth = sns.kdeplot(telco_data_dummies.MonthlyCharges[(telco_data_dummies['Churn']==0)],color = 'Red', shade = True)
mth = sns.kdeplot(telco_data_dummies.MonthlyCharges[(telco_data_dummies['Churn']==1)],ax = mth, color = 'Blue', shade = True)
mth.legend(['No Churn', 'Churn'], loc = 'upper right')
mth.set_ylabel('Density')
mth.set_xlabel('Monthly Charges')
mth.set_title('Monthly Charge by Churn')
plt.show()
```

Churn is high when monthly charge is high

```python
tot = sns.kdeplot(telco_data_dummies.TotalCharges[(telco_data_dummies['Churn']==0)],color = 'Red', shade = True)
tot = sns.kdeplot(telco_data_dummies.TotalCharges[(telco_data_dummies['Churn']==1)],ax = tot, color = 'Blue', shade = True)
tot.legend(['No Churn', 'Churn'], loc = 'upper right')
tot.set_ylabel('Density')
tot.set_xlabel('Total Charges')
tot.set_title('Total Charge by Churn')
plt.show()
```

Higher Churn at lower Total Charges

- Howerver if we combine the insight of 3 parameter i.e. Tenure, Monthly charges, Total charges then the picture is bit clear 
- Higher Monthely charges at lower Tenure result into lower Total charge. Hence all these 3 factor are linked in higher churn

Build a corelation of all predictors with churn 

```python
plt.figure(figsize=(20,8))
telco_data_dummies.corr()['Churn'].sort_values(ascending = False).plot(kind= 'bar')
plt.show()
```

High Churn seen in case Month to Month contracts. No online security. No Tech support, First year of subscription and Fiber Optics Internet

Low Churn is seen case of long term contract. Subscriptions without internet service and the customers engaged for 5+ years 


Factors like Gender, avalibility of PhoneService and Multiple lines have almost no impact on Churn 

```python
plt.figure(figsize=(12,12))
sns.heatmap(telco_data_dummies.corr(), cmap= 'Paired')
plt.show()
```

## Bivariate Analysis

```python
new_df1_target0 = telco_data.loc[telco_data['Churn']==0]
new_df1_target1 = telco_data.loc[telco_data['Churn']==1]
```

```python
def uniplot(df, col, title, hue=None):
    sns.set_style('whitegrid')
    sns.set_context('talk')

    mpl.rcParams['axes.labelsize'] = 20
    mpl.rcParams['axes.titlesize'] = 22
    mpl.rcParams['axes.titlepad'] = 30

    sns.countplot(data=df, x=col, hue=hue)
    plt.title(title)
    plt.show()
```

```python
uniplot(new_df1_target1,col='Partner', title= 'Distribution of gender for Churned customers', hue = 'gender')
```

```python
uniplot(new_df1_target0,col='Partner', title= 'Distribution of gender for non Churned customers', hue = 'gender')
```

```python
uniplot(new_df1_target1,col='PaymentMethod', title= 'Distribution of PaymentMethod for Churned customers', hue = 'gender')
```

```python
uniplot(new_df1_target0,col='PaymentMethod', title= 'Distribution of PaymentMethod for non Churned customers', hue = 'gender')
```

```python
uniplot(new_df1_target1,col='Contract', title= 'Distribution of contract for Churned customers', hue = 'gender')
```

```python
uniplot(new_df1_target1,col='TechSupport', title= 'Distribution of Tech support for Churned customers', hue = 'gender')
```

```python
uniplot(new_df1_target1,col='SeniorCitizen', title= 'Distribution of senior citizen for Churned customers', hue = 'gender')
```
