```python
import pandas as pd 
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import recall_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from imblearn.combine import SMOTEENN
```

```python
df = pd.read_csv('tel_churn.csv')
```

```python
df.head()
```

```python
df = df.drop('Unnamed: 0',axis =1) 
```

```python
df.head()
```

```python
# creating x and y variables 
x = df.drop('Churn', axis =1)
x
```

```python
y = df['Churn']
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.2)
```

Decision Tree Classifier

```python
model_dt = DecisionTreeClassifier(criterion = 'gini', random_state =100, max_depth = 6, min_samples_leaf=8)
model_dt.fit(x_train, y_train)
```

```python
y_pred = model_dt.predict(x_test)
```

```python
y_pred
```

```python
model_dt.score(x_test,y_pred)
```

```python
print(classification_report(y_test, y_pred,labels = [0,1]))
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
sm = SMOTEENN()
x_resampled, y_resampled = sm.fit_resample(x,y)
```

```python
xr_train,xr_test,yr_train,yr_test = train_test_split(x_resampled, y_resampled, test_size=0.2)
```

```python
model_dt_sm = DecisionTreeClassifier(criterion = 'gini', random_state =100, max_depth = 6, min_samples_leaf=8)
model_dt_sm.fit(xr_train, yr_train)
```

```python
yr_pred = model_dt_sm.predict(xr_test)
```

```python
print(classification_report(yr_test, yr_pred,labels = [0,1]))
```

```python
print(confusion_matrix(yr_test, yr_pred))
```

## Randomforest 

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model_rf = RandomForestClassifier(criterion = 'gini', random_state =100, max_depth = 6, min_samples_leaf=8)
model_rf.fit(x_train, y_train)
```

```python
y_pred_rf = model_rf.predict(x_test)
```

```python
print(classification_report(y_test, y_pred_rf,labels = [0,1]))
```

```python
print(confusion_matrix(y_test, y_pred_rf))
```

```python
model_rf_sm = RandomForestClassifier(criterion = 'gini', random_state =100, max_depth = 6, min_samples_leaf=8)
model_rf_sm.fit(xr_train, yr_train)
```

```python
yr_pred_rf = model_rf_sm.predict(xr_test)
```

```python
print(classification_report(yr_test, yr_pred_rf,labels = [0,1]))
```

```python
print(confusion_matrix(yr_test, yr_pred_rf))
```
